### Poverty Rate

```python
NCdata['PovertyRate'] = (NCdata['BelowPovertyE']/NCdata['TotalPovertyE']) *100
GAdata['PovertyRate'] = (GAdata['BelowPovertyE']/GAdata['TotalPovertyE']) *100
```

### Percent Unemployed

```python
NCdata['UnemploymentRate'] = (NCdata['CivilianInLaborforceUnEmployedE']/NCdata['CivilianInLaborforceE']) *100
GAdata['UnemploymentRate'] = (GAdata['CivilianInLaborforceUnEmployedE']/GAdata['CivilianInLaborforceE']) *100
```

### Percent Receiving Social Security

```python
NCdata['PercentReceivingSocialSecurity'] = (NCdata['WithSocialSecurityE']/NCdata['SocialTotalE'])*100
GAdata['PercentReceivingSocialSecurity'] = (GAdata['WithSocialSecurityE']/GAdata['SocialTotalE'])*100
```

### Percent Without Health Insurance

```python
NCdata['NoInsuranceAllAgeGroups'] =  (NCdata['Under19NoInsuranceE'] + NCdata['Age19To34NoInsuranceE']
                                      + NCdata['Age35To64NoInsuranceE'] 
                                      + NCdata['Age65andOverNoInsuranceE'])
GAdata['NoInsuranceAllAgeGroups'] =  (GAdata['Under19NoInsuranceE'] + GAdata['Age19To34NoInsuranceE'] 
                                      + GAdata['Age35To64NoInsuranceE'] 
                                      + GAdata['Age65andOverNoInsuranceE'])
```

```python
NCdata['PercentWithoutHealthInsurance'] = (NCdata['NoInsuranceAllAgeGroups']/NCdata['TotalInsuranceE'])*100
GAdata['PercentWithoutHealthInsurance'] = (GAdata['NoInsuranceAllAgeGroups']/GAdata['TotalInsuranceE'])*100
```

### Percent Single Parent

```python
NCdata['TotalSingleParentsWithChildren'] = (NCdata['MaleNoSpouseChildUnder18E'] + NCdata['FeMaleNoSpouseChildUnder18E'])
GAdata['TotalSingleParentsWithChildren'] = (GAdata['MaleNoSpouseChildUnder18E'] + GAdata['FeMaleNoSpouseChildUnder18E'])
```

```python
NCdata['PercentSingleParent'] =  (NCdata['TotalSingleParentsWithChildren']/NCdata['TotalFamilyE'])*100
GAdata['PercentSingleParent'] =  (GAdata['TotalSingleParentsWithChildren']/GAdata['TotalFamilyE'])*100
```

### Percent Femaled Headed Household

```python
NCdata['PercentFemaleHeaded'] =  (NCdata['FeMaleNoSpouseChildUnder18E']/NCdata['TotalFamilyE'])*100
GAdata['PercentFemaleHeaded'] =  (GAdata['FeMaleNoSpouseChildUnder18E']/GAdata['TotalFamilyE'])*100
```

### Percent Minority

```python
NCdata['TotalMinorities'] = (NCdata['HispanicorLatinoE'] + NCdata['BlackorAfricanAmericanaloneE'] + 
                             NCdata['AmericanIndianandAlaskaNativealoneE'] + NCdata['AsianaloneE'] + 
                            NCdata['NativeHawaiianandOtherPacificIslanderaloneE'] + NCdata['SomeotherracealoneE'] + 
                            NCdata['TwoormoreracesE'])

GAdata['TotalMinorities'] =  (GAdata['HispanicorLatinoE'] + GAdata['BlackorAfricanAmericanaloneE'] + 
                             GAdata['AmericanIndianandAlaskaNativealoneE'] + GAdata['AsianaloneE'] + 
                            GAdata['NativeHawaiianandOtherPacificIslanderaloneE'] + GAdata['SomeotherracealoneE'] + 
                            GAdata['TwoormoreracesE'])
```

```python
NCdata['PercentMinority'] = (NCdata['TotalMinorities'] / NCdata['TotalRaceE'])*100
GAdata['PercentMinority'] = (GAdata['TotalMinorities'] / GAdata['TotalRaceE'])*100
```

### Percent Mobile Homes

```python
NCdata['PercentMobileHomes'] = (NCdata['TotalMobileHomesE']/NCdata['TotalunitsE'])*100
GAdata['PercentMobileHomes'] = (GAdata['TotalMobileHomesE']/GAdata['TotalunitsE'])*100
```

### Percent Renters

```python
NCdata['PercentRenters'] = (NCdata['RenterOccupiedE']/NCdata['TotalTenureE'])*100
GAdata['PercentRenters'] = (GAdata['RenterOccupiedE']/NCdata['TotalTenureE'])*100
```

### Rent Burden 

```python
NCdata['TotalRentBurden'] =  (NCdata['Income20000E'] + NCdata['Income200034999E'] + NCdata['Income35000949999E'] 
                             + NCdata['Income5000074999E'] + NCdata['Income75000E'] + NCdata['NegativeIncomeE'])


GAdata['TotalRentBurden'] =  (GAdata['Income20000E'] + GAdata['Income200034999E'] + GAdata['Income35000949999E'] 
                             + GAdata['Income5000074999E'] + GAdata['Income75000E'] + GAdata['NegativeIncomeE'])
```

```python
NCdata['PercentRentBurden'] =  (NCdata['TotalRentBurden']/ NCdata['TotalRenterUnits30E'])*100

GAdata['PercentRentBurden'] =  (GAdata['TotalRentBurden']/GAdata['TotalRenterUnits30E'])*100
```

### Percent Education Attainment 

```python
NCdata['PercentBachelors'] = (NCdata['BachelorsE']/NCdata['TotalAttainmentE'])*100 
GAdata['PercentBachelors'] = (GAdata['BachelorsE']/GAdata['TotalAttainmentE'])*100 
```

```python
NCdata['PercentNoSchooling'] = (NCdata['NoSchoolingE']/NCdata['TotalAttainmentE'])*100 
GAdata['PercentNoSchooling'] = (GAdata['NoSchoolingE']/GAdata['TotalAttainmentE'])*100 
```
